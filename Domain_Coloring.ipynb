# Domain Coloring

## Introducción

Las formas de representar una función compleja son muy limitadas. Hans Lundmark [Lun04] y Elias Wegert [Weg12] dejan en claro la raz´on por la que se dificulta la visualizaci´on de estas funciones: las funciones complejas viven en un  espacio de cuatro dimensiones. A pesar de ello, se han desarrollado formas para poder visualizarlas. Entre ellas, se encuentran las gráficas de tercera dimensión de la parte real e imaginaria, gráficas del módulo o las transformaciones (estas formas se encuentran ejemplificadas en [UNdEaD]). En esta ocasión, presentaremos la más novedosa y completa de todas, el dominio coloreado, o mejor conocido como domain coloring.


## Domain coloring

Lundmark explica de mantera muy interesante cómo se observan algunas funciones complejas con la técnica y algunas de sus propiedades [Lun04]. También se pueden encontrar otras características de la técnica en [PP09], [Pon18] y en
[WSGD18], quienes tienen una galería virtual de funciones complejas  visualizadas a través del coloreado del dominio.

## Funciones complejas

Definición (Función compleja [pra19]). Sea $S\subset\mathbb{C}$. Una función en variable compleja $f$ definida sobre $S$ es una función en la que a cada número complejo $z\in S$ se le asigna otro número complejo $w\in\mathbb{C}$. El número $w$ es el valor de $f$ sobre $z$, por tanto $w=f(z)$. 
De este modo $f:\mathbb{C}\rightarrow\mathbb{C}.$ Si $z=x+iy$ y $w=u+iv$, entonces, $w=(u,v)=\ u(x,y)+iv(x,y)=\ f(z)=f(x+iy)$. 

Im3.jpg

Este resultado se ve reflejado en la Figura anterior. Como $z$ y $w$ pertenecen al mundo de los números complejos, $f$ tiene dos dimensiones de entrada y dos de salida, ¿cómo representamos en un solo gráfico una función compleja? Veamos por qué no es tan sencillo. Por ejemplo, las funciones $f:\mathbb{R}\rightarrow\mathbb{R}$ tienen una dimensión real de entrada y una de salida, por lo que el plano euclídeo es suficiente para poder representarla gráficamente, por otro lado, las funciones $f:\ \mathbb{R}^2\rightarrow\mathbb{R}$, poseen dos dimensiones reales de entrada y una de salida, de manera que es muy común que se utilice la representación de superficies tridimensionales para su visualización.

Sin embargo, para una función compleja $f:D\ \subset\ \mathbb{C}\rightarrow\mathbb{C}$ se sabe que el gráfico
$$G_f:=(z,f(z))\in \mathbb{C} \times \mathbb{C}:z \in D$$
vive en $\mathbb{C}\times\mathbb{C}$, es decir que tiene cuatro dimensiones reales y aunque dispusiéramos de una herramienta que permitiese su representación gráfica, lastimosamente los sentidos del ser humano perciben únicamente tres dimensiones [weg12] [pon19].

## Coloreado del dominio (en inglés: domain coloring)
Por tanto, si se representa gráficamente una función compleja en el plano $\mathbb{R}^2$, no existirían problemas con la existencia de diferentes proyecciones, pues solo existiría una única representación, de modo que tal representación sería universal. Esta idea se le atribuye a Larry Crone y Hans Lundmark, quienes fueron los primeros en representar funciones complejas, desde finales de la década de 1980, aun así, su técnica no tenía un nombre propio. No fue sino hasta 1998 que Frank Farris realizó una reseña del libro Visual Complex Analysis de Tristan Needham y le asignó el nombre de Domain coloring a la técnica que ya se venía empleando desde Crone y Lundmark [weg12]. 
Dicha técnica se basa en la coloración de los paisajes coloreados, en el que el plano complejo se puede visualizar como una paleta de colores construida bajo algún sistema de coloración. Este procedimiento es mayormente conocido como retrato de fase (ver [weg12]) [pon20].

Las ventajas de los retratos de fase es que son una imagen bidimensional, por lo que no necesitan ser proyectadas y nuestro cerebro está entrenado para interpretar tales imágenes; además, el rango de la fase es pequeña (es un subconjunto del círculo unitario), lo que trae como consecuencia una resolución visual mayor para la fase, permitiendo representar todas las funciones con un mismo esquema de color [weg12].

## Sistemas de coloración

Con respecto al modelo de coloración a utilizar, es preciso recalcar que existen varios y cada uno de ellos permite la visualización de las propiedades de la función que se desea representar de diferente manera. 

### Sistema de coloración HSL
Iniciemos utilizando el modelo HSL (Hue, Saturation, Lightness: Matiz, Saturación, Luminosidad respectivamente), entonces el esquema de coloración col de una función compleja $f:D\rightarrow Y,\ Y\ \subset\mathbb{C}$, está definida por $col:Y\rightarrow HSL$, en la que se asigna a cada elemento $y\in Y$ un color $col(y)$ [pon20]. Recordando la representación polar de un número,  $z=re^{i\theta}$, siendo $\theta$ el argumento de $z$ y $r=\sqrt{x^2+y^2}$ el módulo, con el sistema HSL, el matiz indica la fase y la luminosidad está caracterizada por el módulo:

$$H=Matiz=\theta=Arg\left(z\right),$$
$$L=Luminosidad=r=\sqrt{x^2+y^2}.$$

Además, si se llega a evidenciar que los colores se están concentrando en un punto negro, este punto de concentración se denomina un cero de la función, es decir, el punto donde la función toma el valor de cero; por el contrario, si el punto de concentración es blanco, se denomina polo o singularidad, es decir que este es un punto donde la función no se encuentra definida [weg12] [pon19]. Veamos la siguiente figura, donde se evidencia es el retrato de fase a través del sistema HSL de la función $f(z)=z$.

Im7.jpg

Notemos que los colores rojo, verde y azul aparecen en los ángulos $0, \frac{2\pi}{3} y \frac{4\pi}{3}$ respectivamente, también vemos que los reales positivos se encuentran representados con color rojo, además, en el punto $z=0$, se aprecia un color negro, que indica que este punto es un cero de la función $f(z)$. 

La desventaja con el sistema HSL es que los colores no se muestran muy nítidos, de manera que no es posible identificar de manera precisa el punto que es un cero y/o un polo, por ejemplo, al intentar analizar la función $$f\left(z\right)=\frac{(z-1)}{(z^2+z+1)},$$ sobre la cual se mostró su paisaje analítico coloreado, en este caso no es posible visualizar con precisión dónde se encuentra el cero de la función [weg12] [pon19].

### Sistema HSV

Por eso existe otro sistema que solventa la desventaja de la nitidez en el color que se aprecia mediante el sistema HSL, este es el sistema HSV (\textbf{H}ue, \textbf{S}aturation, \textbf{V}alue: Matiz, Saturación, Valor respectivamente). Aunque tampoco es perfecto, pues al usar este sistema, se pierde la propiedad de poder identificar si el punto donde se acumulan los colores es un cero o un polo, porque no se muestra si este es negro o blanco; se pueden añadir otras herramientas a la representación dada por este sistema para su identificación, las cuales se explicarán más adelante. Veamos a continuación, el retrato de fase simple de la función $f(z) = z$ mediante el sistema HSV:

Im9.jpg

Hasta el momento solo se evidencian los colores que están asignados a cada punto de la función compleja, en los que el argumento del número complejo afecta la cantidad de luz presente en un color, notemos que en los ángulos $0, \frac{2\pi}{3} y \frac{4\pi}{3}$ aparecen los colores primarios; también se puede apreciar que los colores coinciden en el punto $z\ =\ 0$. Al analizar esta función, se puede comprobar que este punto un cero, pero, al mirar su representación gráfica, aún no sabemos qué significa este punto debido a que los retratos de fase simple bajo el modelo HSV no nos brindan más información [pon20].

Ahora bien, si añadimos las curvas de nivel asociadas al módulo, también se pueden añadir las curvas de la fase, incluso, se pueden evidenciar ambas curvas de nivel en un solo gráfico:

Las curvas de nivel del módulo se evidencian en las circunferencias, mientras que las curvas de nivel de la fase son representadas por las rectas. Cuando las circunferencias se ven unas encima de otras en un punto z, nos está indicando que ese punto es un cero de la función, mientras que, si se evidencia una especie de túnel, este punto es un polo. Veamos la representación para $f(z)=\frac{1}{z}$, y notemos que los colores están asignados de manera diferente para cada punto de la función, y aunque pareciera adoptar las mismas curvas de nivel, en este caso, las circunferencias van adoptando una forma de túnel, por lo que el punto $z=0$ que es un cero para $f(z)=z$, es un polo para $f(z)=\frac{1}{z}$:

Donde se puede notar que con el sistema HSV, es más fácil identificar el punto exacto de concentración de colores (independientemente si es un polo o un cero), a comparación del sistema HSL. Aun así, en varias ocasiones podría resultar un poco tedioso la identificación de la acumulación de las circunferencias o el túnel que estas pueden formar. Los ceros y polos son dos elementos esenciales en el estudio de funciones en variable compleja y aunque se pueden calcular de manera analítica, para funciones complejas se encuentra la gran ventaja de que al representarlas por medio de la técnica del dominio coloreado podemos encontrar tales puntos de forma sencilla.

## Domain coloring en python

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import hsv_to_rgb

def coloracion_de_dominios(f, limites=[-2.5, 2.5, -2.5, 2.5], muestras=100, pasos=0.001, grilla=True, tam_rayo=0.0025):

  pasos = [pasos, pasos]
  x = np.arange(limites[0], limites[1], pasos[0])
  y = np.arange(limites[2], limites[3], pasos[1])

  xx, yy = np.meshgrid(x, y)
  z = xx + 1j * yy
  w = f(z)

  img_hsv = obtener_hsv(w, grilla, tam_rayo)
  img_rgb = hsv_to_rgb(img_hsv)

  fig, ax = plt.subplots()
  img = ax.imshow(img_rgb, interpolation='bilinear', extent=limites)
  ax.invert_yaxis()
  
  return fig, ax, img

def obtener_hsv(z, grilla,tam_rayo):
  arg = np.angle(z)
  mod = np.abs(z)
  modificar_mod = _modificar_mod(mod)

  h = (arg / (2 * np.pi)) % 1

  s = modificar_mod.copy()
  s[modificar_mod <  0.5] = 1
  s[modificar_mod >= 0.5] = 1 - (s[modificar_mod >= 0.5] - 0.5) / 5

  v = modificar_mod.copy()
  v[modificar_mod <  0.5] = 0.9 - (0.5 - v[modificar_mod < 0.5]) / 2
  v[modificar_mod >= 0.5] = 0.9

  rayo = np.logical_or(h <= tam_rayo, 1 - tam_rayo <= h)
  s[rayo] = 0.3
  v[rayo] = 1

  for a in np.linspace(0, 1, 12, endpoint=False)[1:]:
    rayo = np.logical_and(a - tam_rayo <= h, h <= a + tam_rayo)
    s[rayo] = 0.3
    v[rayo] = 1

  return np.stack([h, s, v], axis=-1) 

def _modificar_mod(mod):
  mod_max = mod.max()
  mod_min = mod.min()
  mod = mod.copy()
  
  n = 30
  i = 0

  mod_peq = mod <= 1

  while mod_min < 1 and i < n:
    p = np.logical_and(mod_peq, mod <= 1/2)
    m = np.logical_and(mod_peq, np.logical_not(p))
    mod[p] *= 2
    mod_min *= 2
    mod[m] = mod[m] * 2 - 1
    i += 1
    mod_peq = p
  i = 0
  mod_gran = 1 < mod

  while 1 < mod_max and i < n:
    g  = 2 < mod
    m = np.logical_and(mod_gran, np.logical_not(g))
    mod[g] /= 2
    mod_max /= 2
    mod[m] -= 1
    i += 1
    mod_gran = g
  
  return mod

fig, ax, img = coloracion_de_dominios(lambda z: z)
fig, ax, img = coloracion_de_dominios(lambda z: np.sin(z))
plt.show()
```
